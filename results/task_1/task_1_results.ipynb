```python
import pandas as pd
import os
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
plt.style.use("bmh")
import numpy as np
# import plotly.express as px
```

```python
models = os.listdir("data")
len(models)
```

```python
data = pd.concat(
    [pd.read_csv("data/" + m).reset_index(drop = True).assign(model = m[:-4]) for m in tqdm(models)]
)\
.reset_index(drop = True)
print(data.shape)
data.head()
```

```python
plt.figure(figsize=(15, 8))
metric = "f1"
df = data.sort_values("test_" + metric, ascending = True)
plt.plot( df["test_" + metric], df.model, "o-", label = "test")
plt.plot( df["train_" + metric], df.model, "o-", label = "train")
plt.xlabel(metric)
plt.xlim(-.05, 1.05)
plt.xticks(np.linspace(0, 1, 11))
plt.legend()
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
metric = "accuracy"
df = data.sort_values("test_" + metric, ascending = True)
plt.plot( df["test_" + metric], df.model, "o-", label = "test")
plt.plot( df["train_" + metric], df.model, "o-", label = "train")
plt.axvline(.77, ls = "--", color = "black", alpha = 0.5, label = "all_negative")
plt.xlabel(metric)
plt.xlim(-.05, 1.05)
plt.xticks(np.linspace(0, 1, 11))
plt.legend()
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
metric = "precision"
df = data.sort_values("test_" + metric, ascending = True)
plt.plot( df["test_" + metric], df.model, "o-", label = "test")
plt.plot( df["train_" + metric], df.model, "o-", label = "train")
plt.xlabel(metric)
plt.xlim(-.05, 1.05)
plt.xticks(np.linspace(0, 1, 11))
plt.legend()
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
